````r
#loading the neccessary packages
library(tidyverse) #for data wrangling
library(lubridate)  #for data functions
library(ggplot2) #for data visualization
list.files(path = "../input")
````

STEP 1: COLLECT DATA

````r
# importing datasets
m7 <- read.csv("../input/cyclistic-dataset/202207-divvy-tripdata.csv")
m8 <- read.csv("../input/cyclistic-dataset/202208-divvy-tripdata.csv")
m9 <- read.csv("../input/cyclistic-dataset/202209-divvy-publictripdata.csv")
m10 <- read.csv("../input/cyclistic-dataset/202210-divvy-tripdata.csv")
m11 <- read.csv("../input/cyclistic-dataset/202211-divvy-tripdata.csv")
m12<- read.csv("../input/cyclistic-dataset/202212-divvy-tripdata.csv")

m1 <- read.csv("../input/cyclistic-dataset/202301-divvy-tripdata.csv")
m2 <- read.csv("../input/cyclistic-dataset/202302-divvy-tripdata.csv")
m3 <- read.csv("../input/cyclistic-dataset/202303-divvy-tripdata.csv")
m4 <- read.csv("../input/cyclistic-dataset/202304-divvy-tripdata.csv")
m5 <- read.csv("../input/cyclistic-dataset/202305-divvy-tripdata.csv")
m6 <- read.csv("../input/cyclistic-dataset/202306-divvy-tripdata.csv")
````

STEP 2: WRANGLE DATA AND COMBINE INTO A SINGLE FILE

````r
#compare column names of each file for combining
colnames(m7)
colnames(m8)
colnames(m9)
colnames(m10)
colnames(m11)
colnames(m12)
colnames(m1)
colnames(m2)
colnames(m3)
colnames(m4)
colnames(m5)
colnames(m6)
````

````r
# combining into quarters
q1_2023<- rbind(m1,m2,m3)
q2_2023<-rbind(m4,m5,m6)
q3_2022<-rbind(m7,m8,m9)
q4_2022<-rbind(m10,m11,m12)
````

````r
#inspect data frames for irregularities
str(q1_2023)
str(q2_2023)
str(q3_2022)
str(q4_2022)
````

````r
all_trips <- rbind(q3_2022, q4_2022, q1_2023, q2_2023)
````

STEP 3: CLEAN UP AND ADD DATA TO PREPARE FOR ANALYSIS

````r
colnames(all_trips)  #List of column names
````

````r
nrow(all_trips)  #How many rows are in data frame?
````

````r
dim(all_trips)  #Dimensions of the data frame?
````

````r
head(all_trips)  #See the first 6 rows of data frame.
````

````r
str(all_trips)  #See list of columns and data types (numeric, character, etc)
````

````r
summary(all_trips)  #Statistical summary of data. Mainly for numerics
````

````r
table(all_trips$member_casual) # checking for different types of members and their count
````

````r
# creating new column in the dataframe as date, month, day, year, day of week
all_trips$date <- as.Date(all_trips$started_at) #The default format is yyyy-mm-dd
all_trips$month <- format(as.Date(all_trips$date), "%m")
all_trips$day <- format(as.Date(all_trips$date), "%d")
all_trips$year <- format(as.Date(all_trips$date), "%Y")
all_trips$day_of_week <- format(as.Date(all_trips$date), "%A")
````

````r
# calulating the ride length of each ride and inserting into a new column ride_length
all_trips$ride_length <- difftime(all_trips$ended_at,all_trips$started_at)
````

````r
is.numeric(all_trips$ride_length)
````

````r
# converting ride_length to numeric
all_trips$ride_length <- as.numeric(as.character(all_trips$ride_length))
is.numeric(all_trips$ride_length)
````

````r
# checking for irregularities in ride length
table(all_trips$ride_length<0)
````

````r
# removing all rows with ride length <0
all_trips_v2 <- all_trips[!(all_trips$ride_length<0),]
table(all_trips_v2$ride_length<0)
````

STEP 4: CONDUCT DESCRIPTIVE ANALYSIS

````r
# Descriptive analysis on ride_length (all figures in seconds)
mean(all_trips_v2$ride_length) # average (total ride length / rides)
median(all_trips_v2$ride_length) #midpoint number in the ascending array of ride lengths
max(all_trips_v2$ride_length) #longest ride
min(all_trips_v2$ride_length) #shortest ride

summary(all_trips_v2$ride_length)
````

````r
# mean ride length for each type of user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = mean)
````

````r
# median ride length for each type of user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = median)
````

````r
# max ride length for each type of user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = max)
````

````r
# min ride length for each type of user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual, FUN = min)
````

````r
# mean ride length per day of week for each user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
````

````r
# ordering data by days of the week
all_trips_v2$day_of_week <- ordered(all_trips_v2$day_of_week, levels=c("Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"))

# mean ride length per day of week for each user
aggregate(all_trips_v2$ride_length ~ all_trips_v2$member_casual + all_trips_v2$day_of_week, FUN = mean)
````

````r
no_avg_ride <- all_trips_v2 %>% 
  mutate(weekday = wday(started_at, label = TRUE)) %>% 
  group_by(member_casual, weekday) %>% 
  summarise(number_of_rides = n(),average_duration = mean(ride_length)) %>% 
  arrange(member_casual, weekday)

ggplot(data = no_avg_ride, aes(x = weekday, y = number_of_rides, fill = member_casual)) +
  geom_col(position = "dodge")

ggplot(data = no_avg_ride,aes(x = weekday, y = average_duration, fill = member_casual)) +
    geom_col(position = "dodge")
````

*For member users during* ***weekdays*** *the* ***average duration is almost same*** *as well as* ***total no. of rides fluctuate with Wednesday's having the highest traction***,
*during* ***weekends*** *the* ***average duration slightly increases*** and ***total no. of rides see a decrease***

*For casual users the* ***average duration during the weekdays fluctuate but remains lower weekends*** *and* ***total no. of rides remains around the same***,
*during* ***weekends*** *the* ***average duration slightly increases*** *and* ***total no. of rides see a small increase on Sunday's and a big bump on Saturday's.***

*No. of rides is higher for Members whereas average duration is much higher for casual users.*

````r
avg_ride_length=all_trips_v2 %>% group_by(member_casual,month) %>%  
    summarise(avg_duration = mean(ride_length)) %>% 
    arrange(member_casual, month)
    
# avg ride duration by month bar chart 
ggplot(data =avg_ride_length,aes(x = month, y = avg_duration, fill = member_casual)) +
    geom_col(position = "dodge")

#avg ride duration by month line graph 
ggplot(data =avg_ride_length,aes(x = month, y = avg_duration,group = member_casual)) +
    geom_line(aes(colour = member_casual)) + 
    geom_point()
````

*Overall* ***average duration for casual users is much higher than member users***, *both type of users have their* ***lowest*** *around* ***December, January, February and March***

````r
all_trips_v2 %>% group_by(member_casual) %>%  
summarise(number_of_rides = n())

#no. of rides by month bar chart 
all_trips_v2 %>% group_by(member_casual,month) %>%  
summarise(number_of_rides = n()) %>% 
arrange(member_casual, month) %>% 
ggplot(aes(x = month, y = number_of_rides, fill = member_casual)) +
geom_col(position = "dodge")

#no. of rides by month line graph
all_trips_v2 %>% 
group_by(month,member_casual) %>%  
summarise(number_of_rides = n()) %>% 
arrange(month,member_casual) %>% 
ggplot(aes(x = month, y = number_of_rides,group = member_casual)) +
geom_line(aes(colour = member_casual)) + 
geom_point()
````

***Member users*** *have* ***higher no. of rides as compared to casual users*** *and no. of rides start* ***increasing around April until July, August*** *before starting to slowly decline.*

***November, December, January, February and March have the lowest no. of rides for casual users***, *even lower than the lowest month for member users (December)*

````r
# per rideable type by member percentage
all_trips_v2 %>% 
    group_by(rideable_type,member_casual) %>%  
    summarise(number_of_ride = n()) %>%
    mutate(percentage = (number_of_ride / sum(number_of_ride)) * 100)%>%
    ggplot(aes(x = rideable_type, y = percentage, fill = member_casual)) +
        geom_col(position = "dodge")
````

*Close to* ***60%*** *of* ***classic and electric bikes*** *are used by* ***member users***

````r
# lets calculate the type of bike used most by causal riders and members
all_trips_v2 %>% 
  mutate(bike_type = all_trips_v2$rideable_type) %>% 
  group_by(member_casual, bike_type) %>% 
  summarise(number_of_rides = n()) %>%
          arrange(member_casual, bike_type)  %>% 
  ggplot(aes(x = member_casual,y = number_of_rides, fill = bike_type)) +
  geom_col(position = "dodge")
````

***Electric bikes are used more than Classic bikes by both casual and member users and docked bikes are only used by casual users***
